## Executive Summary

Across 32 traders and 211,224 trades (May 2023 – May 2025), three findings stand out:

1. **Traders make more money during Fear than Greed.** Average PnL per closed trade is **$126 in Fear regimes vs $46 in Extreme Greed** — a 2.7× gap. The Mann-Whitney U test confirms this is highly significant (p < 0.001).
2. **The top 5 traders are aggressively contrarian.** They go **97% LONG during Extreme Fear** and only **12% LONG during Extreme Greed** — buying panic and selling euphoria. The bottom 5 do the opposite (perma-short, even during fear rallies) and lose money.
3. **Sentiment value negatively correlates with daily PnL** (Spearman ρ = -0.165, p < 0.001). As the market gets greedier, profits shrink. This is consistent with mean-reversion alpha being strongest when the crowd is most directional.

**Strategy implication:** The dataset supports a contrarian framework — fade extreme sentiment, especially extreme greed (where bottom traders lose the most). Top trader behavior provides a template worth studying further.

## 1. Setup and Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100

SENTIMENT_ORDER = ['Extreme Fear', 'Fear', 'Neutral', 'Greed', 'Extreme Greed']
SENTIMENT_COLORS = {
    'Extreme Fear': '#8B0000', 'Fear': '#CD5C5C', 'Neutral': '#808080',
    'Greed': '#90EE90', 'Extreme Greed': '#006400'
}
```

```python
# Load both datasets
trades = pd.read_csv('data/historical_data.csv')
sentiment = pd.read_csv('data/fear_greed_index.csv')

print(f'Trader data: {len(trades):,} rows, {trades["Account"].nunique()} unique accounts')
print(f'Sentiment data: {len(sentiment):,} rows')
trades.head(3)
```

## 2. Data Quality Notes

- **Trader dataset:** 211,224 trades from 32 unique accounts across 246 coins, May 2023 → May 2025. No missing values.
- **Sentiment dataset:** 2,644 daily observations from Feb 2018 → May 2025, classified into 5 buckets.
- **Column mismatch from task description:** The task email lists a `leverage` column, but **the CSV does not contain it**. All leverage-related analysis is therefore excluded. The `event` column from the task is actually called `Direction` (values: `Open Long`, `Close Long`, `Open Short`, `Close Short`, etc.) — used as the event type throughout.
- **PnL convention:** `Closed PnL` is only populated on close events (50% of rows are opens with PnL=0 by design — these are excluded from PnL analysis).
- **Curated sample:** 32 unique traders is a small, curated set rather than random retail. Findings describe these specific traders' behavior, not the broader market.

## 3. Cleaning and Merge

```python
# Parse timestamps (format DD-MM-YYYY HH:MM)
trades['datetime'] = pd.to_datetime(trades['Timestamp IST'], format='%d-%m-%Y %H:%M')
trades['date'] = trades['datetime'].dt.date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date

# Merge on date
df = trades.merge(sentiment[['date', 'value', 'classification']], on='date', how='left')
df = df.rename(columns={'value': 'fg_value', 'classification': 'sentiment'})

unmatched = df['sentiment'].isna().sum()
print(f'Trades without sentiment match: {unmatched} ({100*unmatched/len(df):.3f}%)')

df = df.dropna(subset=['sentiment'])
df['sentiment'] = pd.Categorical(df['sentiment'], categories=SENTIMENT_ORDER, ordered=True)

# Helper subsets
closes = df[df['Direction'].str.contains('Close', case=False, na=False)].copy()
closes['win'] = closes['Closed PnL'] > 0
opens = df[df['Direction'].str.contains('Open', case=False, na=False)].copy()
opens['position_type'] = opens['Direction'].apply(
    lambda x: 'Long' if 'Long' in x else ('Short' if 'Short' in x else 'Other'))

print(f'Total trades after merge: {len(df):,}')
print(f'Close events (for PnL analysis): {len(closes):,}')
print(f'Open events (for direction analysis): {len(opens):,}')
```

## 4. Insight 1 — Win Rate by Market Sentiment

**Question:** Do traders win more often during Fear or Greed periods?

A 'win' is defined as a close event with positive realized PnL.

```python
win_rate = closes.groupby('sentiment', observed=True)['win'].agg(['mean', 'count'])
win_rate['mean'] = win_rate['mean'] * 100
win_rate.columns = ['Win Rate %', 'Trade Count']
win_rate.round(2)
```

**Finding:** Win rate is highest in **Fear** (88.6%) and **Extreme Greed** (87.4%) regimes, and lowest in normal **Greed** (76.1%). The U-shape suggests traders perform best when sentiment is one-sided enough to be predictable. The middle-Greed regime is the worst — likely because direction is ambiguous and traders get whipsawed.

## 5. Insight 2 — Realized PnL by Sentiment

**Question:** Beyond just winning, do traders make more *money* in any regime?

```python
pnl_summary = closes.groupby('sentiment', observed=True)['Closed PnL'].agg(['mean', 'median', 'sum', 'count'])
pnl_summary.columns = ['Mean PnL ($)', 'Median PnL ($)', 'Total PnL ($)', 'Trade Count']
pnl_summary.round(2)
```

```python
# Statistical test
fear_pnl = closes[closes['sentiment'].isin(['Extreme Fear', 'Fear'])]['Closed PnL']
greed_pnl = closes[closes['sentiment'].isin(['Greed', 'Extreme Greed'])]['Closed PnL']
u_stat, p_val = stats.mannwhitneyu(fear_pnl, greed_pnl, alternative='two-sided')
print(f'Mann-Whitney U test (Fear vs Greed PnL):')
print(f'  Fear days mean: ${fear_pnl.mean():.2f}  |  Greed days mean: ${greed_pnl.mean():.2f}')
print(f'  U = {u_stat:,.0f}, p = {p_val:.2e}')
```

**Finding:** Mean PnL per trade is **$126 during Fear** vs **$46 during Extreme Greed** — a 2.7× gap. Mann-Whitney U confirms this is highly significant (p ≈ 0). Of the $7.3M total PnL across all 32 traders, **$3.35M (46%)** was made during plain 'Fear' regimes alone — far more than any other bucket despite Fear days representing only 29% of trades.

This is the core finding: *the most profitable trading environment for this cohort is moderate fear, not greed.*

## 6. Insight 3 — Long vs Short Bias by Sentiment

**Question:** Does the cohort flip its positioning based on sentiment?

```python
bias = opens.groupby(['sentiment', 'position_type'], observed=True).size().unstack(fill_value=0)
bias['Long %'] = 100 * bias['Long'] / (bias['Long'] + bias['Short'])
bias.round(2)
```

**Finding:** The cohort exhibits a **clear contrarian directional bias**: 69% long during Extreme Fear, dropping to 42% long during normal Greed. The sentiment-to-positioning relationship is monotonic — as the market gets greedier, the cohort gets relatively more short. This is consistent with experienced traders fading extreme sentiment.

## 7. Insight 4 (Headline) — Top vs Bottom Traders

**Question:** Among the 32 traders, what specifically separates the winners from the losers?

Define Top 5 / Bottom 5 by total realized PnL across the full dataset.

```python
trader_pnl = closes.groupby('Account')['Closed PnL'].agg(['sum', 'count', 'mean'])
trader_pnl.columns = ['Total PnL', 'Trade Count', 'Mean PnL']
trader_pnl = trader_pnl.sort_values('Total PnL', ascending=False)

top5 = trader_pnl.head(5).index.tolist()
bottom5 = trader_pnl.tail(5).index.tolist()

print(f'Top 5 traders total PnL: ${trader_pnl.head(5)["Total PnL"].sum():,.0f}')
print(f'Bottom 5 traders total PnL: ${trader_pnl.tail(5)["Total PnL"].sum():,.0f}')
print(f'\nTop 5 vs Bottom 5 long-bias by sentiment:')

def get_ls_pct(opens_df):
    g = opens_df.groupby('sentiment', observed=True)['position_type'].value_counts().unstack(fill_value=0)
    g['Long %'] = 100 * g['Long'] / (g['Long'] + g['Short'])
    return g['Long %'].reindex(SENTIMENT_ORDER)

top_ls = get_ls_pct(opens[opens['Account'].isin(top5)])
bot_ls = get_ls_pct(opens[opens['Account'].isin(bottom5)])
comp = pd.DataFrame({'Top 5 Long %': top_ls, 'Bottom 5 Long %': bot_ls}).round(1)
comp
```

**Finding (headline):** The difference between winners and losers in this dataset is *almost entirely directional bias by sentiment*. Top traders are **97% long during Extreme Fear** and drop to **12% long during Extreme Greed** — a textbook contrarian profile. Bottom traders are perma-short across every sentiment regime (8–13% long), which means they bet against the rebounds from fear and got crushed.

Average PnL per trade for the bottom 5 turns **strongly negative during Greed (−$281) and Extreme Greed (−$568)** while top traders remain positive across all sentiments. This is the single highest-signal finding in the dataset: *position direction conditioned on sentiment explains most of the performance dispersion.*

## 8. Insight 5 — Daily Aggregate PnL vs Sentiment Score

```python
daily = closes.groupby('date').agg(
    total_pnl=('Closed PnL', 'sum'),
    fg_value=('fg_value', 'first')).reset_index()

r_p, p_p = stats.pearsonr(daily['fg_value'], daily['total_pnl'])
r_s, p_s = stats.spearmanr(daily['fg_value'], daily['total_pnl'])
print(f'Pearson:  r = {r_p:.4f} (p = {p_p:.4f})')
print(f'Spearman: r = {r_s:.4f} (p = {p_s:.4f})')
```

**Finding:** Daily total PnL is **negatively correlated** with the Fear & Greed value (Spearman ρ = −0.17, p < 0.001). The effect is small in absolute terms but statistically robust, and it points the same direction as every other finding: as sentiment gets greedier, this cohort's profits shrink.

## 9. Trading Strategy Implications

1. **The cohort's edge is contrarian and concentrated in Fear regimes.** Strategies that long during fear and short during greed (especially extreme greed) match the historical winner pattern.
2. **Greed periods are the danger zone for unsophisticated traders.** Bottom traders lose almost all their money during Greed and Extreme Greed. Risk management (smaller size, tighter stops) should escalate as sentiment approaches greed extremes.
3. **Positioning is more predictive than entry timing.** The dispersion between top and bottom traders is explained almost entirely by *what direction* they were betting, not *when*. A simple rule — 'don't be short during Extreme Fear' — would have moved several bottom-5 traders out of the loss column.
4. **The middle-Greed bucket is uniquely treacherous** (76% win rate, lowest of all regimes). Direction is ambiguous and the cohort gets whipsawed. Position-sizing models could weight signals lower in this regime.

## 10. Limitations

- **Causation vs correlation:** Sentiment is partly a function of recent price action; we cannot claim sentiment *causes* trader outcomes — only that they co-move.
- **Sample size:** 32 traders is a small curated set, not a random retail sample. Findings may not generalize.
- **No leverage data:** The task description mentioned leverage, but it is absent from the CSV. Leverage-conditioned analysis (e.g., liquidation rates by sentiment) is therefore not possible.
- **Survivorship/selection:** Unknown whether the 32 accounts were selected for being notable. Caveat all findings accordingly.
- **No regime-shift handling:** The Fear/Greed buckets are point-in-time; transitions between regimes are not modeled.

## 11. Conclusion

The single most important finding in this dataset is that *trader performance dispersion is dominated by directional bias conditioned on sentiment*. Top traders fade the crowd; bottom traders stay perma-short and bleed during rallies. The Fear regime is where most of the profit lives ($3.35M of $7.3M total). The simplest actionable insight is also the strongest: **be long during Fear, neutral-to-short during Greed, and especially cautious during Extreme Greed where the worst outcomes cluster.**
